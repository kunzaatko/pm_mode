```python
import numpy as np
import pandas as pd
import sys
sys.path.append('..')
from models.feature_extraction.feature_extraction import Data
```

```python
data = Data()
data.LL_data = pd.read_csv('../LL_data.csv', index_col=0)
data.SL_data = pd.read_csv('../SL_data.csv', index_col=0)
data.matches = pd.read_csv('../matches.csv', parse_dates=['Date','Open'], index_col=0)
data.match_data = pd.read_csv('../match_data.csv', parse_dates=['Date'],index_col=0)
```

```python
productivity = pd.DataFrame(index=data.matches.index)
for i in [1,2,5,10]:
    productivity['HID_goals_scored_to_match_#_m#'+str(i)] = data.matches.apply(lambda row: data.goals_to_match_ratio(row.HID,number_of_matches=i,MatchID=row.name,goal_type='scored'), axis=1)
    productivity['HID_goals_conceded_to_match_#_m#'+str(i)] = data.matches.apply(lambda row: data.goals_to_match_ratio(row.HID,number_of_matches=i,MatchID=row.name,goal_type='conceded'), axis=1) 
    productivity['AID_goals_scored_to_match_#_m#'+str(i)] = data.matches.apply(lambda row: data.goals_to_match_ratio(row.AID,number_of_matches=i,MatchID=row.name,goal_type='scored'), axis=1)
    productivity['AID_goals_conceded_to_match_#_m#'+str(i)] = data.matches.apply(lambda row: data.goals_to_match_ratio(row.AID,number_of_matches=i,MatchID=row.name,goal_type='conceded'), axis=1)                                                                            
```

```python
productivity[['A','H']] = data.matches[['A','H']]
```

```python
productivity[[col for col in productivity.columns if 'HID' in col and 'scored' in col] + ['H']].corr()
```

```python
productivity[[col for col in productivity.columns if 'AID' in col and 'scored' in col] + ['A']].corr()
```

Jak je vidět, tak korelace je největší pro vstřelené góly za posledních 10 zápasů a to $0.088218$ pro `HID` a $0.116055$ pro `AID`

```python
productivity[[col for col in productivity.columns if 'HID' in col and 'conceded' in col] + ['A']].corr()
```
