```python
#funkce k předpovědí výsledku
def compare(expected): 

    if (expected[0] - expected[1]) > 0.05:
        return np.array([1, 0 ,0])

    elif (expected[1] - expected[0]) > 0.05:
        return np.array([0, 0 ,1])
    else:
        return np.array([0, 1, 0])
```

```python
predict = np.zeros((len(df), 3))
i=0
for row in expected:
        predict[i, :] = compare(row)
        i=i+1
```
