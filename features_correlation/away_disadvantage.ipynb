```python
import numpy as np
import pandas as pd
import sys
sys.path.append('..')
from models.feature_extraction.feature_extraction import Data
```

```python
data = Data()
data.LL_data = pd.read_csv('../LL_data.csv', index_col=0)
data.SL_data = pd.read_csv('../SL_data.csv', index_col=0)
data.matches = pd.read_csv('../matches.csv', parse_dates=['Date','Open'], index_col=0)
data.match_data = pd.read_csv('../match_data.csv', parse_dates=['Date'],index_col=0)
```

`away_disadvantage` is $\frac{P(\text{lose}\mid \text{away})}{P(\text{lose})}$. That is the deterioration of the team when playing away. The correlation represents the influence of the teams worse performance in the away matches than average lose rate overall on its losing in a away match.

```python
away_advantages = pd.DataFrame(index=data.matches.index, dtype=np.float64)
A_side_dis = pd.Series(index=data.matches.index)
A_side_dis = data.matches.apply(lambda row: data.away_disadvantage(row.AID, MatchID=row.name, method='all'),axis=1)

away_advantages['A_side_dis_rate_sur'] = A_side_dis.apply(lambda row: row[0])
away_advantages['A_side_dis_rate_rat'] = A_side_dis.apply(lambda row: row[1])
away_advantages['A_side_dis_rate'] = A_side_dis.apply(lambda row: row[2])
```

```python
away_advantages['A'] = data.matches.A + data.matches.D * .5
```

```python
away_advantages.corr()
```

`A_side_dis_rate_sur` is the Away disadvantage rate surplus $\ldots$ $P(\text{lose} | \text{away}) - P(\text{lose})$

`A_side_dis_rate_rat` is the Away disadvantage rate ratio $\ldots$ $\frac{P(\text{lose} | \text{away})}{P(\text{lose})} $

`A_side_dis_rate` is the Away disadvantage rate $\ldots$ $P(\text{lose} | \text{away})$
