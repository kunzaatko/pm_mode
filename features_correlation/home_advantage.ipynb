```python
import numpy as np
import pandas as pd
import sys
sys.path.append('..')
from models.feature_extraction.feature_extraction import Data
```

```python
data = Data()
data.LL_data = pd.read_csv('../LL_data.csv', index_col=0)
data.SL_data = pd.read_csv('../SL_data.csv', index_col=0)
data.matches = pd.read_csv('../matches.csv', parse_dates=['Date','Open'], index_col=0)
data.match_data = pd.read_csv('../match_data.csv', parse_dates=['Date'],index_col=0)
```

`home_advantage` is $\frac{P(\text{win}\mid \text{home})}{P(\text{win})}$. That is the improvement of the team when playing home. The correlation represents the influence of the teams better play in the home matches than average win rate overall on its winning in a home match.

```python
home_advantages = pd.DataFrame(index=data.matches.index, dtype=np.float64)
H_side_adv = pd.Series(index=data.matches.index)
H_side_adv = data.matches.apply(lambda row: data.home_advantage(row.AID, MatchID=row.name, method='all'),axis=1)
```

```python
home_advantages['H_side_adv_rate_sur'] = H_side_adv.apply(lambda row: row[0])
home_advantages['H_side_adv_rate_rat'] = H_side_adv.apply(lambda row: row[1])
home_advantages['H_side_adv_rate'] = H_side_adv.apply(lambda row: row[2])
```

```python
home_advantages['H'] = data.matches.H + data.matches.D *.5
```

```python
home_advantages.corr()
```

`H_side_adv_rate_sur` is the Home advantage rate surplus $\ldots$ $P(\text{win} | \text{home}) - P(\text{win})$

`H_side_adv_rate_rat` is the Home advantage rate ratio $\ldots$ $\frac{P(\text{win} | \text{home})}{P(\text{win})} $

`H_side_adv_rate` is the Home advantage rate $\ldots$ $P(\text{win} | \text{home})$
