```python
import numpy as np
import pandas as pd
import sys
sys.path.append('..')
from models.feature_extraction.feature_extraction import Data
```

```python
data = Data()
data.LL_data = pd.read_csv('../LL_data.csv', index_col=0)
data.SL_data = pd.read_csv('../SL_data.csv', index_col=0)
data.matches = pd.read_csv('../matches.csv', parse_dates=['Date','Open'], index_col=0)
data.match_data = pd.read_csv('../match_data.csv', parse_dates=['Date'],index_col=0)
```

```python
A_rate = data.matches.apply(lambda row: data.away_disadvantage(row.AID, MatchID=row.name, method='rate'),axis=1)
H_rate = data.matches.apply(lambda row: data.home_advantage(row.HID, MatchID=row.name, method='rate'),axis=1)
```

```python
(H_rate + A_rate).corr((data.matches.H + data.matches.D * .5))
```

`H_rate + A_rate` is $P(\text{win} \mid \text{home})_{\text{HID}} + P(\text{lose} \mid \text{away})_{\text{AID}}$
